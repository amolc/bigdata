# Module 1: PySpark Examples

This notebook contains runnable examples corresponding to Chapters 1 through 5.

## Chapter 3: Spark Setup and Environment

Initialize a local Spark session.

```
from pyspark.sql import SparkSession

# Create a SparkSession
spark = SparkSession.builder \
    .appName("Module1_Examples") \
    .master("local[*]") \
    .getOrCreate()

print("Spark version:", spark.version)
```

## Chapter 4: Core Spark Concepts

Working with RDDs and defining schemas.

```
# RDD Example
data = [1, 2, 3, 4, 5]
rdd = spark.sparkContext.parallelize(data)
rdd_mapped = rdd.map(lambda x: x * 2)
print("Mapped RDD:", rdd_mapped.collect())
```

```
# Schema Definition
from pyspark.sql.types import StructType, StructField, StringType, IntegerType

schema = StructType([
    StructField("Name", StringType(), True),
    StructField("Age", IntegerType(), True),
    StructField("City", StringType(), True)
])

people_data = [("Alice", 28, "New York"), ("Bob", 25, "San Francisco")]
df = spark.createDataFrame(people_data, schema=schema)
df.show()
```

## Chapter 5: Spark SQL

DataFrame operations and SQL queries.

```
# DataFrame Filtering
df_filtered = df.filter(df.Age > 26)
df_filtered.show()
```

```
# Temporary Views and SQL
df.createOrReplaceTempView("people")

sql_df = spark.sql("SELECT City, AVG(Age) as AvgAge FROM people GROUP BY City")
sql_df.show()
```

## Chapter 6: Data Processing

Aggregations, UDFs, and Dates.

```
# UDF Example
from pyspark.sql.functions import udf, col
from pyspark.sql.types import StringType

def age_category(age):
    return "Adult" if age >= 18 else "Minor"

age_udf = udf(age_category, StringType())
df_with_category = df.withColumn("Category", age_udf(col("Age")))
df_with_category.show()
```

```
# Dates and Timestamps
from pyspark.sql.functions import current_date, date_add

df_dates = df.withColumn("Today", current_date()) \
             .withColumn("Next_Week", date_add(col("Today"), 7))
df_dates.select("Name", "Today", "Next_Week").show()
```
